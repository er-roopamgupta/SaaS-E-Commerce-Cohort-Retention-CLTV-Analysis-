```python
import pandas as pd

# Load and prepare data
df = pd.read_csv('Ecommerce_Transactions_Clean.csv')
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], dayfirst=True)

df['Cohort_Month'] = df.groupby('Customer_ID')['Purchase_Date'].transform('min').dt.to_period('M')
df['Purchase_Month'] = df['Purchase_Date'].dt.to_period('M')

df['Cohort_Index'] = (df['Purchase_Month'].dt.year - df['Cohort_Month'].dt.year) * 12 + \
                     (df['Purchase_Month'].dt.month - df['Cohort_Month'].dt.month)

# Cohort Matrix for Total Revenue by Acquisition Channel
revenue_cohort = df.groupby(['Acquisition_Channel', 'Cohort_Index'])['Total_Amount'].sum().unstack(fill_value=0)

print("--- Total Revenue Matrix by Acquisition Channel ---")
print(revenue_cohort)
```
