```python
import pandas as pd
```

```python
# Filter for VIP segments in specific states

sales_data = pd.read_csv(r"C:\Users\Sakshi Pawar\Downloads\Ecommerce_Transactions_Clean.csv")
target_states = ['Maharashtra', 'Delhi', 'West Bengal', 'Tamil Nadu']
vip_performance = sales_data[
    (sales_data['Customer_Segment'] == 'VIP') & 
    (sales_data['State'].isin(target_states))
]

# Aggregate data by State and Gender
vip_summary = vip_performance.groupby(['State', 'Gender']).agg(
    Total_Spent=('Total_Amount', 'sum'),
    Avg_Quantity=('Quantity', 'mean'),
    Transaction_Count=('Transaction_ID', 'count')
).reset_index()

vip_summary
```
