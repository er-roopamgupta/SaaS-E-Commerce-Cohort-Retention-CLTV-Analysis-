```python
import pandas as pd
```

```python
# Load data and convert date column to datetime format
df = pd.read_csv("Ecommerce_Transactions_Clean.csv")
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# Named Query 1: Calculate the Cohort Month for every individual transaction
df['Cohort_Month'] = df.groupby('Customer_ID')['Purchase_Date'].transform('min').dt.to_period('M')

# Named Query 2: Get total new unique users acquired per Cohort Month
cohort_counts = df.groupby('Cohort_Month')['Customer_ID'].nunique().reset_index(name='New_Users')

# Preview the results
print("--- TRANSACTIONS WITH COHORT MONTH ---")
print(df[['Customer_ID', 'Purchase_Date', 'Cohort_Month']].head())

print("\n--- NEW USERS ACQUIRED PER COHORT ---")
print(cohort_counts)
```
