```python
import pandas as pd

sales_data = pd.read_csv(r"C:\Users\Sakshi Pawar\Downloads\Ecommerce_Transactions_Clean.csv")

sales_data['Purchase_Date'] = pd.to_datetime(sales_data['Purchase_Date'])

sales_data['Month'] = sales_data['Purchase_Date'].dt.to_period('M')

monthly_sales = sales_data.groupby('Month')['Total_Amount'].sum().reset_index()

monthly_sales
```
