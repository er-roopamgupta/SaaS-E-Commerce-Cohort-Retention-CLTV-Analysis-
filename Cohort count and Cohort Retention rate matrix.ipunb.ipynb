```python
import pandas as pd

# Load data and setup cohort features (Steps 1–4)
df = pd.read_csv('Ecommerce_Transactions_Cleaned.csv')
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

df['Cohort_Month'] = df.groupby('Customer_ID')['Purchase_Date'].transform('min').dt.to_period('M')
df['Purchase_Month'] = df['Purchase_Date'].dt.to_period('M')

df['Cohort_Index'] = (df['Purchase_Month'].dt.year - df['Cohort_Month'].dt.year) * 12 + \
                     (df['Purchase_Month'].dt.month - df['Cohort_Month'].dt.month)

# Step 5: Pivot the data to create the Cohort Retention Matrix
cohort_matrix = df.groupby(['Cohort_Month', 'Cohort_Index'])['Customer_ID'].nunique().unstack(fill_value=0)

# View absolute active customer count per cohort over time
print("--- Cohort Count Matrix ---")
print(cohort_matrix)

# Calculate retention rate as a percentage of Month 0 cohort size
cohort_size = cohort_matrix.iloc[:, 0]
retention_matrix = cohort_matrix.divide(cohort_size, axis=0).round(4) * 100

print("\n--- Cohort Retention Rate Matrix (%) ---")
print(retention_matrix)
```
