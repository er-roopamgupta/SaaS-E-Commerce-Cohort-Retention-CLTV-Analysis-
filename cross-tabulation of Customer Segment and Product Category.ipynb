```python
import pandas as pd
```

```python
# Cross-tabulate geographic states with the acquisition channel
sales_data = pd.read_csv('Ecommerce_Transactions_Clean.csv')

state_channel = pd.crosstab(
    sales_data['State'],
    sales_data['Acquisition_Channel'],
    normalize='index'
) * 100

# Display formatted percentages
state_channel.round(2).astype(str) + '%'
```

```python
# Create cross-tabulation of Customer Segment and Product Category
sales_data = pd.read_csv('Ecommerce_Transactions_Clean.csv')


segment_category = pd.crosstab(
    sales_data['Customer_Segment'],
    sales_data['Product_Category'],
    normalize='index'
) * 100

# Display formatted percentages
segment_category.round(2).astype(str) + '%'
```
