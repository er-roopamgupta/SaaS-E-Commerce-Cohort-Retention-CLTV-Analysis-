```python
import pandas as pd

sales_data = pd.read_csv('Ecommerce_Transactions_Clean.csv')


# Calculate the percentage breakdown of Payment_Status across Customer_Segments
payment_health = pd.crosstab(
    sales_data['Customer_Segment'], 
    sales_data['Payment_Status'], 
    normalize='index'
) * 100

print("Payment Status Distribution (%) by Customer Segment:")
display(payment_health.round(2))
```
