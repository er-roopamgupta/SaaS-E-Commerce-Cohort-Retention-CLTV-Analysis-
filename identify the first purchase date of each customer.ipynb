```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv('Ecommerce_Transactions_Clean.csv')

# 2. Convert the Purchase_Date column to datetime format
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# 3. Identify the first purchase date for each customer (Month 0)
cohort_df = df.groupby('Customer_ID')['Purchase_Date'].min().reset_index()

# 4. Format the date to Year-Month format to define the Cohort Month
cohort_df['Cohort_Month'] = cohort_df['Purchase_Date'].dt.to_period('M')

# Display the resulting Month 0 mappings
print(cohort_df[['Customer_ID', 'Cohort_Month']].head())
```
