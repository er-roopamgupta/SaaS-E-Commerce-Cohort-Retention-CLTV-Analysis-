```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv('Ecommerce_Transactions_Clean.csv')

# 2. Filter for successful transactions & convert date
successful_df = df[df['Payment_Status'] == 'Success'].copy()
successful_df['Purchase_Date'] = pd.to_datetime(successful_df['Purchase_Date'])

# 3. Get the first sale date for each product category
category_df = successful_df.groupby('Product_Category')['Purchase_Date'].min().reset_index()

# 4. Format date to Period
category_df['Launch_Month'] = category_df['Purchase_Date'].dt.to_period('M')

# Display result
print(category_df[['Product_Category', 'Launch_Month']])
```
